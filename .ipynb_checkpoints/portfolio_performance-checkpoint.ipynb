```python
import numpy as np
import datetime as dt
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt
```

## get data

```python
stocks = [
    'TSLA', # tesla
    'EUNL.DE',  # core msci world i shares
    'BNO', # United States Brent Oil Fund,
    'ISTB', # iShares Core 1-5 Year USD Bond ETF
]
```

```python
endDate = dt.datetime.now()
endDate
```

```python
startDate = endDate - dt.timedelta(days=365)
startDate
```

```python
stockData = pdr.get_data_yahoo(stocks, startDate, end=endDate)
```

## display data

```python
stockData.head(5)
```

## extract only Close

```python
stockCloseData = stockData['Close']
```

```python
stockCloseData.head(3)
```

## calc returns

```python
returns = stockCloseData.pct_change()
returns.head(5)
```

## calc mean returns

```python
mean_returns = returns.mean()
mean_returns.head(5)
```

## calc covariance matrix

```python
cov_matrix = returns.cov()
cov_matrix
```

```python
plt.matshow(cov_matrix)
plt.show()
```

## portfolio performance

```python
def portfolioPerformance(weights, meanReturns, covMatrix):
    actual_returns = np.sum(meanReturns*weights)*252
    std = np.sqrt(
            np.dot(weights.T,np.dot(covMatrix, weights))
           )*np.sqrt(252)
    return actual_returns, std
```

```python
weights = np.array([0.1, 0.5, 0.2, 0.2])
weights
```

```python
portfolio_returns, portfolio_std = portfolioPerformance(weights, mean_returns, cov_matrix)
```

```python
print(portfolio_returns, portfolio_std)
```

```python
print(round(portfolio_returns*100, 2), round(portfolio_std*100, 2))
```
